```python
path = model.cost_complexity_pruning_path(X_train,y_train)
ccp_alphas, impurities =path.ccp_alphas, path.impurities
```

```python
ccp_alphas
```

```python
models=[]
for ccp_alpha in ccp_alphas:
    model=DecisionTreeClassifier(random_state=0,ccp_alpha=ccp_alpha)
    model.fit(X_train, y_train)
    models.append(model)
print(" Number of trees in the last tree is: {} with ccp_alpha value: {} ".format(models[-1].tree_.node_count, ccp_alphas[-1]))
```

```python
train_score=[model.score(X_train,y_train) for model in models]
test_score=[model.score(X_test,y_test) for model in models]

fig , ax=plt.subplots()
ax.set_xlabel('aplha')
ax.set_ylabel('accuracy')
ax.set_title('Accuracy vs alpha for train and test sets')
ax.plot(ccp_alphas, train_score, marker ="o" , label='train' , drawstyle="steps-post")
ax.plot(ccp_alphas , test_score , marker ="o",label='test',drawstyle ="steps-post")
ax.legend()
plt.show()
```

when ccp_alpha is set to zero by keeping default parameters for class: DecisiontreeClassifier, the tree overfits, leading 100% of accuracy for train sets and 94% accuracy for test sets. As alpha increases more of the tree is pruned thus creating a decision tree that generalizes better. n this example, setting ccp_alpha=0.05 maximizes the testing accuracy.

```python
model = DecisionTreeClassifier(random_state=0, ccp_alpha=0.05)
model.fit(X_train,y_train)
```

```python
pred=model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```
